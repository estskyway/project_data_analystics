```python
import pandas as pd
import pymongo as mg
from mecab import MeCab
mecab = MeCab()
df_reviews = pd.read_csv('../../datasets/minjeong_mongoDB.csv')
df_reviews[:1]
```

```python
df_reviews = df_reviews['reviews'].tolist()
df_reviews[:3]
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
text = ' '.join(df_reviews[:1000])  # df_reviews[:100]의 각 요소들을 공백으로 구분하여 하나의 문자열로 결합
morphs_list = okt.morphs(text)  # 토큰화
print(morphs_list)
```

```python
okt.pos(df_reviews[0]) #형태소 분석
```

```python
result = okt.nouns(text) # 명사만 출력 
print(result)
```

```python
morphemes = okt.morphs(text) #조사만 출력 
all_particles = [morpheme for morpheme in morphemes if okt.pos(morpheme)[0][1] == 'Josa']
print(all_particles)
```

### 불용어 제거

```python
# 중복된 조사 제거
unique_morphemes = list(set(morphemes))
```

```python
# 불용어 처리할 조사 목록
all_particles = ['은','이','가','하고', '로서','도','으로', '에', '을', '께서', '에서', '라고', '이며', '으로도', '에서는', '까지', '까지도', '말고는', '라고는', '이랑', '에게', '에만', '한테', '엔', '에서도', '보단', '이고', '이라서', '랑', '으로는', '이에요', '밖엔', '하고는']  # 모든 조사를 추가하세요
```

```python
# 모든 조사를 불용어로 처리하여 제외
filtered_morphemes = [morpheme for morpheme in morphemes if morpheme not in all_particles]
```

```python
# 불용어 개수 계산
num_stopwords = len(morphemes) - len(filtered_morphemes)

# 결과 출력
print("불용어 개수:", num_stopwords)
print("명사만 추출된 형태소:", filtered_morphemes)
```

```python
print(filtered_morphemes)
```

### 한단어만 추출 

```python
# 명사만 추출된 형태소에서 중복을 제거하고, 한 단어만 추출
unique_nouns = list(set(filtered_morphemes))

print("명사만 추출된 형태소 중 중복을 제거한 결과:")
print(unique_nouns)
```

```python
# 명사만 추출된 형태소에서 한 글자짜리 명사 추출
oneword_list = []
for oneword in filtered_morphemes:
    if len(oneword) == 1:
        oneword_list.append(oneword)

print("한 단어만 추출된 명사:")
print(oneword_list)
```
