```python
import pandas as pd
df = pd.read_csv('../../datasets/hyein_mongoDB.csv')
df[:2]
```

```python
df_review = df['text']
# df_review[:3].tolist()
review_list = df_review.tolist()
review_list[:5]
```

### 정규표현식
- 공백: \s, 공백이 아닌 값: \S
- 숫자: \d, 숫자가 아닌 값: \D
- 제외: ^
- 알파벳, 숫자: \w

```python
import re
pattern = r'[\s]'
text_regex = re.sub(pattern=pattern, repl='', string=review_list[0])
text_regex
```

```python
from mecab import MeCab
mecab = MeCab()
```

```python
# 형태소 분석
mecab.morphs(sentence=review_list[0])
```

```python
# 품사 달기
mecab.pos(review_list[0])
```

```python
for review in review_list : 
    morphs_list = mecab.morphs(review)
    print(morphs_list)
```

### 불용어 제거
- 문장을 토큰화 후 조사, 접속사 등을 제거

```python
stopwords = ['은', '는', '이', '가', '도', '께', '서']
```

### 불용어를 제거하여 형태소 분석

```python
nouns=[]
for review in review_list : 
    for noun in mecab.nouns(review):
        if noun not in stopwords :
            nouns.append(noun)
nouns
```
