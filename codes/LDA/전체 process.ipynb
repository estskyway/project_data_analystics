#### 형태소 분석

```python
from mecab import MeCab
mecab = MeCab()
```

```python
# 한 문장으로 만들기
reviews = ' '.join(reviews_list)
reviews
```

```python
# 토큰화
mecab.morphs(reviews)
```

```python
# 품사달기
morphs = mecab.pos(reviews)
morphs
```

```python
# 토큰화 함수
def tokenize_reviews_list(reviews):
    tokens = mecab.morphs(reviews)
    return ' '.join(tokens)
```

```python
df['tokenized_review'] = df['reviews'].apply(tokenize_reviews_list)
```

```python
words = ' '.join(reviews_list)
words
```

```python
stopwords = df_stopwords['stopword'].values
onewords = df_onewords['one_char_keyword'].values

def tokenizer(raw, stopword=stopwords, oneword=onewords):
    result_list = list()
    try : 
        for word in mecab.nouns(raw) :
            if word not in stopword:
                result_list.append(word)
            # if word in stopword :
            #     result_lsit.append(word)
    except:
        pass
    return ' '.join(result_list)
```

```python
df['reviews'].apply(tokenizer)
```
