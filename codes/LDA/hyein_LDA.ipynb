### LDA분석

```python
import pandas as pd
```

```python
df = pd.read_csv('../../datasets/total_data.csv')
df
```

```python
# 필요없는 컬럼 삭제
df = df.drop(columns=['Unnamed: 0', 'stars', 'rating'], axis=1)
df[:5]
```

```python
# null값 확인
df.info()
```

```python
df = df.dropna()
df[:5]
```

```python
# 긍정 / 부정 classification컬럼 분리
positive_list = list(df[df['classification'] == '긍정']['classification'])
negative_list = list(df[df['classification'] == '부정']['classification'])
```

```python
len(positive_list), len(negative_list)
```

```python
# 전체 데이터
preprocess_data = df['preprocess_review'].dropna().str.split().tolist()
```

```python
# 긍정 / 부정 리뷰 데이터 분리
positive_data = df[df['classification'] == '긍정']['preprocess_review'].str.split().tolist()
negative_data = df[df['classification'] == '부정']['preprocess_review'].str.split().tolist()
```

### transform with dictionary

```python
from gensim.corpora import Dictionary
from gensim.models import LdaModel
```

```python
dictionary = Dictionary(preprocess_data)
dictionary
```

```python
dictionary_positive = Dictionary(positive_data)
dictionary_positive
```

```python
dictionary_negative = Dictionary(negative_data)
dictionary_negative
```

```python
dictionary_positive.token2id
```

```python
dictionary_negative.token2id
```

```python
# transform(긍정 1개)
dictionary_positive.doc2bow(positive_data[0])
```

```python
# transform(부정 1개)
dictionary_negative.doc2bow(negative_data[0])
```

```python
# transform(긍정 전체)
corpus_list_positive = list()
for sentence in positive_data:
    morphs_vector = dictionary_positive.doc2bow(sentence)
    corpus_list_positive.append(morphs_vector) # 리스트에 morphs_vector값을 하나씩 담기
# corpus_list_positive
```

```python
# transform(부정 전체)
corpus_list_negative = list()
for sentence in negative_data:
    morphs_vector = dictionary_negative.doc2bow(sentence)
    corpus_list_negative.append(morphs_vector) # 리스트에 morphs_vector값을 하나씩 담기
# corpus_list_negative
```

```python
# fit(긍정)
lda_model_positive = LdaModel(corpus=corpus_list_positive, id2word=dictionary_positive, num_topics=3)
lda_model_positive
```

```python
# fit(부정)
lda_model_negative = LdaModel(corpus=corpus_list_negative, id2word=dictionary_negative, num_topics=3)
lda_model_negative
```

### 최적의 토픽 개수 - 긍정

```python
from gensim.models.coherencemodel import CoherenceModel
```

```python
start_topic = 2
end_topic = 10
coherence_scores_positive = []
perplexity_scores_positive = []
for topic_number in range(start_topic, end_topic+1):
    lda_model = LdaModel(corpus=corpus_list_positive, id2word=dictionary_positive, num_topics=topic_number)  # fit
    coherenceModel = CoherenceModel(model=lda_model, texts=positive_data, dictionary=dictionary_positive) # 인스턴스화
    coherence_scores_positive.append(coherenceModel.get_coherence())  # 일관성 점수
    perplexity_scores_positive.append(lda_model.log_perplexity(corpus_list_positive))  # 혼잡도 점수
```

```python
df_score_positive = pd.DataFrame([coherence_scores_positive, perplexity_scores_positive]).T
df_score_positive

# 일관성 점수가 높고 혼잡도 점수가 낮은 것을 선택
# index=0이 최적의 토픽 개수이므로 2개
```

### 최적의 토픽 개수 - 부정

```python
start_topic = 2
end_topic = 10
coherence_scores_negative = list()
perplexity_scores_negative = list()
for topic_number in range(start_topic, end_topic+1):
    lda_model = LdaModel(corpus=corpus_list_negative, id2word=dictionary_negative, num_topics=topic_number)  # fit
    coherenceModel = CoherenceModel(model=lda_model_negative, texts=negative_data, dictionary=dictionary_negative) #인스턴스화
    coherence_scores_negative.append(coherenceModel.get_coherence())  # 일관성 점수를 list에 담음
    perplexity_scores_negative.append(lda_model.log_perplexity(corpus_list_negative))  # 혼잡도 점수를 list에 담음
```

```python
df_score_negative = pd.DataFrame([coherence_scores_negative, perplexity_scores_negative]).T
df_score_negative

# 일관성 점수가 높고 혼잡도 점수가 낮은 것을 선택
# index= 0이 최적의 토픽 개수이므로  2개
```
