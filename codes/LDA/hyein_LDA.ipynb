### LDA분석

```python
import pandas as pd
```

```python
df = pd.read_csv('../../datasets/total_data.csv')
df
```

```python
# 필요없는 컬럼 삭제
df = df.drop(columns=['Unnamed: 0', 'stars', 'rating'], axis=1)
df[:5]
```

```python
# null값 확인
df.info()
```

```python
df = df.dropna()
df[:5]
```

```python
# 긍정 / 부정 classification컬럼 분리
positive_list = list(df[df['classification'] == '긍정']['classification'])
negative_list = list(df[df['classification'] == '부정']['classification'])
```

```python
len(positive_list), len(negative_list)
```

```python
# 긍정 / 부정 리뷰 데이터 분리
positive_data = df[df['classification'] == '긍정']['preprocess_review'].str.split().tolist()
negative_data = df[df['classification'] == '부정']['preprocess_review'].str.split().tolist()
```

### transform with dictionary

```python
from gensim.corpora import Dictionary
from gensim.models import LdaModel
```

```python
dictionary_positive = Dictionary(positive_data)
dictionary_positive
```

```python
dictionary_negative = Dictionary(negative_data)
dictionary_negative
```

```python
dictionary.token2id
```

```python
# transform(긍정 1개)
dictionary.doc2bow(positive_data[0])
```

```python
# transform(부정 1개)
dictionary.doc2bow(negative_data[0])
```

```python
# transform(긍정 전체)
corpus_list_positive = list()
for sentence in positive_data:
    morphs_vector = dictionary.doc2bow(sentence)
    corpus_list_positive.append(morphs_vector) # 리스트에 morphs_vector값을 하나씩 담기
```

```python
# transform(부정 전체)
corpus_list_negative = list()
for sentence in negative_data:
    morphs_vector = dictionary.doc2bow(sentence)
    corpus_list_negative.append(morphs_vector) # 리스트에 morphs_vector값을 하나씩 담기
```

```python
# fit(긍정)
lda_model_positive = LdaModel(corpus=corpus_list_positive, id2word=dictionary_positive, num_topics=3)
```

```python
# fit(부정)
lda_model_negative = LdaModel(corpus=corpus_list_negative, id2word=dictionary_negative, num_topics=3)
```
