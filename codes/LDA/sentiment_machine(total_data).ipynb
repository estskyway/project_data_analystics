### sentiment_machnine
- 전체 데이터 교육시킨 감성분석을 위한 머신러닝

```python
import pandas as pd
```

```python
df = pd.read_csv('../../datasets/total_data.csv')
df[:5]
```

```python
# 필요없는 컬럼 삭제
df = df.drop(columns=['Unnamed: 0'], axis=1)
df[:5]
```

```python
df.info()
```

#### 벡터 토큰화(단어 숫자화)

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
reviews_list = list(df['reviews'])
reviews_list[:5]
```

```python
# 교육(학습)
countVectorizer = CountVectorizer()
countVectorizer.fit(reviews_list)
```

```python
countVectorizer.vocabulary_
```

```python
tfidfVectorizer = TfidfVectorizer(use_idf=True)
tfidfVectorizer.fit(reviews_list) # 2차원 배열로 교육
```

#### 서비스 배포(TfidfVectorizer)

```python
import pickle
with open('../../datasets/sentiment_TfidfVectorizer.pkl', 'wb') as TfidfVectorizer_file:
    pickle.dump(obj=tfidfVectorizer, file=TfidfVectorizer_file)
    pass
```

```python
tfidfVectorizer.idf_
```

```python
# 벡터화
features = tfidfVectorizer.transform(reviews_list) # 학습 시 2차원 이용했으므로 변환도 2차원 적용
features
```

#### 목표변수 생성

```python
target = list(df['classification'])
target[:5]
```

#### 머신러닝 재학습

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(features, target)
```

#### 서비스 배포(머신러닝)

```python
with open('../../datasets/sentiment_machine.pkl', 'wb') as sentiment_machine_file:
    pickle.dump(obj=model, file=sentiment_machine_file)
    pass
```
